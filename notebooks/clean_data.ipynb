```python
folder = r"C:\Users\kanam\OneDrive\Desktop\jb\data"

loads   = pd.read_csv(os.path.join(folder, "loads.csv"))
routes  = pd.read_csv(os.path.join(folder, "routes.csv"))
trips   = pd.read_csv(os.path.join(folder, "trips.csv"))
drivers = pd.read_csv(os.path.join(folder, "drivers.csv"))
```

```python
df = df.merge(loads,   on="load_id",   how="left")
df = df.merge(routes,  on="route_id",  how="left")
df = df.merge(trips,   on="trip_id",   how="left")
df = df.merge(drivers, on="driver_id", how="left")
```

```python
cols = [
    "on_time_flag",
    "detention_minutes",
    "weight_lbs",
    "typical_distance_miles",
    "base_rate_per_mile",
    "years_experience"
]
data = df[cols].copy()
data = data.dropna()
```
